```python
df = pd.read_csv('/content/gdrive/MyDrive/Plane_ticket/Data_Train.csv')
df
```

```python
df.dropna(inplace = True)
df.drop_duplicates(keep='first',inplace=True)
df.head()
```

```python
x_train = pd.DataFrame(df.drop('Price',axis=1))
y_train = pd.DataFrame(df['Price'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x_train,y_train,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
X_train['Duration'] = X_train['Duration'].str.replace("h",'*60').str.replace(" ", "+").str.replace('m','*1').apply(eval) 
X_test['Duration'] = X_test['Duration'].str.replace("h",'*60').str.replace(" ", "+").str.replace('m','*1').apply(eval) 
```

```python
X_train["Journey_day"] = X_train['Date_of_Journey'].str.split('/').str[0].astype(int)
X_train["Journey_month"] = X_train['Date_of_Journey'].str.split('/').str[1].astype(int)
X_train.drop(["Date_of_Journey"], axis = 1, inplace = True)
X_test["Journey_day"] = X_test['Date_of_Journey'].str.split('/').str[0].astype(int)
X_test["Journey_month"] = X_test['Date_of_Journey'].str.split('/').str[1].astype(int)
X_test.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
X_train["Dep_hour"] = pd.to_datetime(X_train["Dep_Time"]).dt.hour
X_train["Dep_min"] = pd.to_datetime(X_train["Dep_Time"]).dt.minute
X_train.drop(["Dep_Time"], axis = 1, inplace = True)
X_test["Dep_hour"] = pd.to_datetime(X_test["Dep_Time"]).dt.hour
X_test["Dep_min"] = pd.to_datetime(X_test["Dep_Time"]).dt.minute
X_test.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
X_train["Arrival_hour"] = pd.to_datetime(X_train.Arrival_Time).dt.hour
X_train["Arrival_min"] = pd.to_datetime(X_train.Arrival_Time).dt.minute
X_train.drop(["Arrival_Time"], axis = 1, inplace = True)
X_test["Arrival_hour"] = pd.to_datetime(X_test.Arrival_Time).dt.hour
X_test["Arrival_min"] = pd.to_datetime(X_test.Arrival_Time).dt.minute
X_test.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
train_categorical_data = X_train.select_dtypes(exclude=['int64', 'float','int32'])
train_numerical_data = X_train.select_dtypes(include=['int64', 'float','int32'])

test_categorical_data = X_test.select_dtypes(exclude=['int64', 'float','int32','int32'])
test_numerical_data  = X_test.select_dtypes(include=['int64', 'float','int32'])
train_categorical_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_categorical_data = train_categorical_data.apply(LabelEncoder().fit_transform)
test_categorical_data = test_categorical_data.apply(LabelEncoder().fit_transform)
train_categorical_data.head()
```

```python
X = pd.concat([train_categorical_data, train_numerical_data], axis=1)
Y = Y_train['Price']
test_set = pd.concat([test_categorical_data, test_numerical_data], axis=1)
X
```

```python
model = XGBClassifier()
model.fit(X,Y)
```

```python
Y_pred = model.predict(test_set)
predictions = [(value) for value in Y_pred]
```

```python
accuracy = accuracy_score(Y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
